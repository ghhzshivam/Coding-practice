### Hashing

Hashing is a method for sorting and indexing data. we do this by creating keys created using formulas.

Time complexity best - O(1)/worst - Q(N) {Based on collisions}

Hash Function - used to map arbitary size of data to fixed size.

Key: Input data by a user

Hash value: value returned by hash function.

Hash Table: It is a DS which implements an associative array abstract data type, a structure that can map key to values.

Collision: When two different keys to a hash function produce same values.


Properties of good hash function.

It dustributes hash values uniformely across the hash tables.

It has to use all input data


#### Collision Resolution techniques

1) Direct Chaining: Implements Buckets as a Linked list colliding elements are stored in this lists. Hash table full problem never arise.

2) Open Addressing: Colliding elements are stored in a open bucketa. During storage and lookup these are found so called probing.
    
    a) Linear probing: It places new key into closest following empty cell.
    
    b) Quadratic probing: Addign arbitrary quadratic polynomial to the index until an empty cell is found. eg 1^2, 2^2, 3^2.
        So if  if hash function gives 10 so 10 + 1^2 if not empty so 10 + 2^2  etc.
    
    c) Double Hashing: Interval between probes is computed by another hash function.

Pros and Cons of collision resolution techniques.

Direct Chaining:

Pros: Hash table never gets full.

Cons: Huge linked list causes performance leaks(Time complexity for search becomes O(n)).

Open Addressing:

Pros: Easy implementation.

Cons: When Hash Table is full, creating a new hash table affects performance(Time complexity search operation become O(N))


##### If input size is known we always use "Open Addressing"
    
##### If we performe deleteion operation frequently we use "Direct chaining"

Usecase: Password Verification, File system(File path is mapped to physical location on disk)
